```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from scipy import stats

df = pd.read_csv(r"dataset_2.csv")
```

```python
plt.hist(df['BMI'], bins=10, edgecolor='black')
plt.xlabel('BMI')
plt.ylabel('Frequency')
# plt.title('Distribution of BMI')

# plt.xticks(np.arange(df['BMI'].min(), df['BMI'].max(), step=1))

plt.show()
```

```python
# Scatterplot: BMI vs Daily Screen Time
plt.figure(figsize=(8, 6))
sns.regplot(x='DailyScreenTimeHours', y='BMI', data=df, ci=None, scatter_kws={'alpha':0.5, 'edgecolor':'black'}, line_kws={'color':'red', 'linewidth':2})
plt.xlabel('Daily Screen Time (hours)', fontsize=14)
plt.ylabel('BMI', fontsize=14)
# plt.title('BMI vs Daily Screen Time', fontsize=16)
plt.legend(handles=[plt.Line2D([0], [0], color='red', linewidth=2)], labels=['Regression line'])
plt.grid(True, alpha=0.3)
plt.tight_layout()
plt.show()
```

```python
# Scatterplot: BMI vs Fast Food Meals
plt.figure(figsize=(8, 6))
sns.regplot(x='FastFoodMealsPerWeek', y='BMI', data=df, ci=None, scatter_kws={'alpha':0.5, 'edgecolor':'black'}, line_kws={'color':'red', 'linewidth':2})
plt.xlabel('Fast food meals (weekly)', fontsize=14)
plt.ylabel('BMI', fontsize=14)
# plt.title('BMI vs Fast Food Meals per Week', fontsize=16)
plt.legend(handles=[plt.Line2D([0], [0], color='red', linewidth=2)], labels=['Regression line'])
plt.grid(True, alpha=0.3)
plt.tight_layout()
plt.show()
```

```python
# Scatterplot: BMI vs Weekly Exercise
plt.figure(figsize=(8, 6))
sns.regplot(x='WeeklyExerciseHours', y='BMI', data=df, ci=None, scatter_kws={'alpha':0.5, 'edgecolor':'black'}, line_kws={'color':'red', 'linewidth':2})
plt.xlabel('Weekly Exercise (Hours)', fontsize=14)
plt.ylabel('BMI', fontsize=14)
# plt.title('BMI vs Weekly Exercise', fontsize=16)
plt.legend(handles=[plt.Line2D([0], [0], color='red', linewidth=2)], labels=['Regression line'])
plt.grid(True, alpha=0.3)
plt.tight_layout()
plt.show()
```

```python
# Box plot: BMI by Gender with color variation and customized line properties
plt.figure(figsize=(8, 6))
sns.boxplot(data=df, x='Gender', y='BMI', palette='Set2', width=0.5, linewidth=1)
plt.xlabel('Gender', fontsize=14)
plt.ylabel('BMI', fontsize=14)
plt.xticks([0, 1], ['Male', 'Female'], fontsize=12)
# plt.title('BMI Distribution by Gender')
plt.show()
#awdadawdawd
```
