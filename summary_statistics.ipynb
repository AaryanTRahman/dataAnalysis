# Solar Panel Data Analysis - Summary Statistics

This notebook loads the solar panel data and generates summary statistics using pandas.

## Load and Explore the Data

Load the CSV file into a pandas DataFrame and display the first few rows to understand the data structure.

```python
import pandas as pd

# Load the CSV file
df = pd.read_csv('STA201 Project - Sheet1.csv')

# Display the first few rows
df.head()
```

## Generate Summary Statistics with describe()

Use the pandas `describe()` method to generate summary statistics (count, mean, std, min, max, quartiles) for all numeric columns.

```python
# Generate summary statistics for all numeric columns
df.describe()
```

## Summary Statistics by Panel Type

Group the data by PanelType and generate summary statistics separately for Monocrystalline and Polycrystalline panels.

```python
# Summary statistics grouped by Panel Type
df.groupby('PanelType').describe()
```

## Box Plot: Mono vs Poly Output Voltage

Generate a box plot diagram comparing the output voltages of Monocrystalline and Polycrystalline panels.

```python
import matplotlib.pyplot as plt

# Create box plot for Mono and Poly output voltages
fig, ax = plt.subplots(figsize=(8, 6))

# Prepare data - drop NaN values for each voltage type
mono_voltage = df['Mono Voltage'].dropna()
poly_voltage = df['Poly Voltage'].dropna()

# Create box plot
data = [mono_voltage, poly_voltage]
labels = ['Monocrystalline', 'Polycrystalline']
bp = ax.boxplot(data, tick_labels=labels, patch_artist=True, widths=0.5)

# Customize the plot
ax.set_title('Box and Whiskers Plot', fontsize=14, color="#666666")

# Set tick label font size
ax.tick_params(axis='both', labelsize=11)

ax.grid(axis='y', alpha=0.3)

# Add color to boxes and median lines
colors = ["#184FE6", "#8d3aec"]

# Color the box patches
for patch, color in zip(bp['boxes'], colors):
    patch.set_facecolor(color)
    patch.set_edgecolor('black')
    patch.set_linewidth(1)
    patch.set_alpha(0.7)

# Color all whiskers and caps to black
for whisker in bp['whiskers']:
    whisker.set_color('black')
    whisker.set_linewidth(1)

for cap in bp['caps']:
    cap.set_color('black')
    cap.set_linewidth(1)

# Color the median lines to black
for median in bp['medians']:
    median.set_color('black')
    median.set_linewidth(1)

plt.tight_layout()
plt.show()
```
